Ridge Regression

```python
from sklearn.linear_model import Ridge
import pandas as pd

# Function for building and backtesting the model
def build_and_backtest_model(city_weather, target_col, alpha=0.1, start=365, step=90):
    predictors = city_weather.columns[~city_weather.columns.isin(["target", "totalsnow_cm", 'precipitation_target', 'uvindex_target', 'mintempc_target'])]

    model = Ridge(alpha=alpha)
    all_predictions = []

    for i in range(start, city_weather.shape[0], step):
        train = city_weather.iloc[:i, :]
        test = city_weather.iloc[i:(i + step), :]

        model.fit(train[predictors], train[target_col])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test[target_col], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)

    return pd.concat(all_predictions)

# Function to print predictions for a specific city and target variable
def print_predictions(city, target_col, precipmm, maxtempc, uvindex, mintempc):
    print(f"\nPredictions for {city} - {target_col}:")
    print(f"\nPredictions for {city} - {target_col}:")

    print(f"\nPredictions for {city} - precipmm:")
    print(precipmm)

    print(f"\nPredictions for {city} - maxtempc:")
    print(maxtempc)

    print(f"\nPredictions for {city} - uvindex:")
    print(uvindex)

    print(f"\nPredictions for {city} - mintempc:")
    print(mintempc)

# Example usage for Pune's weather data
target_col = "precipmm"  # specify the target variable
predictions_pune_precipmm = build_and_backtest_model(weather_pune, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_pune_maxtempc = build_and_backtest_model(weather_pune, target_col, alpha=0.1)
target_col = "uvindex"
predictions_pune_uvindex = build_and_backtest_model(weather_pune, target_col, alpha=0.1)
target_col = "mintempc"
predictions_pune_mintempc = build_and_backtest_model(weather_pune, target_col, alpha=0.1)

# Example usage for Bombay's weather data
target_col = "precipmm"  # specify the target variable
predictions_bombay_precipmm = build_and_backtest_model(weather_bombay, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_bombay_maxtempc = build_and_backtest_model(weather_bombay, target_col, alpha=0.1)
target_col = "uvindex"
predictions_bombay_uvindex = build_and_backtest_model(weather_bombay, target_col, alpha=0.1)
target_col = "mintempc"
predictions_bombay_mintempc = build_and_backtest_model(weather_bombay, target_col, alpha=0.1)

# Example usage for Kanpur's weather data
target_col = "precipmm"  # specify the target variable
predictions_kanpur_precipmm = build_and_backtest_model(weather_kanpur, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_kanpur_maxtempc = build_and_backtest_model(weather_kanpur, target_col, alpha=0.1)
target_col = "uvindex"
predictions_kanpur_uvindex = build_and_backtest_model(weather_kanpur, target_col, alpha=0.1)
target_col = "mintempc"
predictions_kanpur_mintempc = build_and_backtest_model(weather_kanpur, target_col, alpha=0.1)

# Example usage for Nagpur's weather data
target_col = "precipmm"  # specify the target variable
predictions_nagpur_precipmm = build_and_backtest_model(weather_nagpur, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_nagpur_maxtempc = build_and_backtest_model(weather_nagpur, target_col, alpha=0.1)
target_col = "uvindex"
predictions_nagpur_uvindex = build_and_backtest_model(weather_nagpur, target_col, alpha=0.1)
target_col = "mintempc"
predictions_nagpur_mintempc = build_and_backtest_model(weather_nagpur, target_col, alpha=0.1)

# Example usage for Jaipur's weather data
target_col = "precipmm"  # specify the target variable
predictions_jaipur_precipmm = build_and_backtest_model(weather_jaipur, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_jaipur_maxtempc = build_and_backtest_model(weather_jaipur, target_col, alpha=0.1)
target_col = "uvindex"
predictions_jaipur_uvindex = build_and_backtest_model(weather_jaipur, target_col, alpha=0.1)
target_col = "mintempc"
predictions_jaipur_mintempc = build_and_backtest_model(weather_jaipur, target_col, alpha=0.1)

# Example usage for Delhi's weather data
target_col = "precipmm"  # specify the target variable
predictions_delhi_precipmm = build_and_backtest_model(weather_delhi, target_col, alpha=0.1)
target_col = "maxtempc"
predictions_delhi_maxtempc = build_and_backtest_model(weather_delhi, target_col, alpha=0.1)
target_col = "uvindex"
predictions_delhi_uvindex = build_and_backtest_model(weather_delhi, target_col, alpha=0.1)
target_col = "mintempc"
predictions_delhi_mintempc = build_and_backtest_model(weather_delhi, target_col, alpha=0.1)
```

Calculation of Mean Absolute Error & Accuracy

```python
from sklearn.metrics import mean_absolute_error

def calculate_metrics(predictions, target_col):
    # Calculate Mean Absolute Error
    mae = mean_absolute_error(predictions["actual"], predictions["prediction"])
    
    # Calculate the mean of prediction values
    mean_actual = predictions["actual"].mean()
    
    # Calculate accuracy in percentage
    accuracy_percentage = 100 * (1 - mae / mean_actual)
    
    return mae, accuracy_percentage

# Calculate metrics for maxtempc
mae_maxtempc, accuracy_maxtempc = calculate_metrics(selected_predictions_maxtempc, "maxtempc")
print(f"Mean Absolute Error for maxtempc in {selected_city}: {mae_maxtempc}")
print(f"Accuracy for maxtempc in {selected_city}: {accuracy_maxtempc:.2f}%\n")

# Calculate metrics for precipmm
mae_precipmm, accuracy_precipmm = calculate_metrics(selected_predictions_precipmm, "precipmm")
print(f"Mean Absolute Error for precipmm in {selected_city}: {mae_precipmm}")
print(f"Accuracy for precipmm in {selected_city}: {accuracy_precipmm:.2f}%\n")

# Calculate metrics for uvindex
mae_uvindex, accuracy_uvindex = calculate_metrics(selected_predictions_uvindex, "uvindex")
print(f"Mean Absolute Error for uvindex in {selected_city}: {mae_uvindex}")
print(f"Accuracy for uvindex in {selected_city}: {accuracy_uvindex:.2f}%\n")

# Calculate metrics for mintempc
mae_mintempc, accuracy_mintempc = calculate_metrics(selected_predictions_mintempc, "mintempc")
print(f"Mean Absolute Error for mintempc in {selected_city}: {mae_mintempc}")
print(f"Accuracy for mintempc in {selected_city}: {accuracy_mintempc:.2f}%\n")
```
